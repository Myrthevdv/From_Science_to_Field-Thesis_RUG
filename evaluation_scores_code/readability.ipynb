```python
import pandas as pd
import textstat
from google.colab import files
```

```python
df_output = pd.read_excel('Summarized_Articles_Phi3.xlsx')
```

```python
readability_scores_summarized_articles_df = pd.DataFrame()

for index, row in df_output.iterrows():
    title = row['Title']
    readability_scores_summarized_articles_df.loc[index, 'Title'] = title

    for col_name in df_output.columns[1:]:
        summary = df_output.loc[index, col_name]
        flesch_reading_ease_score = textstat.flesch_reading_ease(summary)
        flesch_kincaid_grade_score = textstat.flesch_kincaid_grade(summary)
        readability_scores_summarized_articles_df.loc[index, f'{col_name} flesch_reading_ease_score '] = float(flesch_reading_ease_score)
        readability_scores_summarized_articles_df.loc[index, f'{col_name} flesch_kincaid_grade_score '] = float(flesch_kincaid_grade_score)
```

```python
readability_scores_summarized_articles_df.to_excel("Readability_scores_summarized_articles_df.xlsx")
```
